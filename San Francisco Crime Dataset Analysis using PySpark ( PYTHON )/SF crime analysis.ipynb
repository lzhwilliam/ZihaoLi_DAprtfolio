# **PART II: DATA PREPERATION**

```
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql import functions as F
from pyspark.sql.functions import year, month, to_date, to_timestamp, hour



spark = SparkSession \
    .builder\
    .appName("crime analysis") \
    .config("spark.some.config.option", "some-value") \
    .getOrCreate()

df_opt1 = spark.read.format("csv").option("header", "true").load(data_path)

df_opt1.createOrReplaceTempView("sf_crime")



df_opt1 = df_opt1.withColumn('Date', to_date(df_opt1.Date, "MM/d/yyyy"))
df_opt1 = df_opt1.withColumn('Time', to_timestamp(df_opt1.Time))
df_opt1 = df_opt1.withColumn('Hour', hour(df_opt1['Time']))
df_opt1 = df_opt1.withColumn('Year', year(df_opt1.Date))
df_opt1 = df_opt1.withColumn('month', month(df_opt1.Date))

df_opt1 = df_opt1.withColumn('X', df_opt1.X.cast('float'))
df_opt1 = df_opt1.withColumn('Y', df_opt1.Y.cast('float'))

display(df_opt1)
```

**EDA PART III:NUMBER OF CRIMES ON EACH SUNDAY OVER THE YEARS**

```
from pyspark.sql.functions import udf
from pyspark.sql.types import BooleanType

sunday = df_opt1.filter(df_opt1.DayOfWeek == 'Sunday')

def sfdowntown(latitude, longitude):
    if latitude >= -122.4313 and latitude <= -122.4213 and longitude >= 37.7540 and longitude <= 37.7740:
      return True
    return False


sundayCrime_udf = udf(sfdowntown, BooleanType())
sunday_crime_df = sunday.withColumn('isDowntown', sundayCrime_udf('X', 'Y'))
sunday_crime_df = sunday_crime_df.filter(sunday_crime_df.isDowntown == True)
#display(sunday_crime_df)
total_count = sunday_crime_df.count()
print('San Francisco crime total counts on sunday over the year is ', total_count)
```

```
sunday.createOrReplaceTempView("sunday")
count = spark.sql("select Date, count(*) as count from sunday group by Date")

count.createOrReplaceTempView("count")
avg_count = spark.sql("select round(avg(count), 0) as avg_count from count")
print("average count crimes happened on sunday is around ")
display(avg_count)
```

**EDA PART VII: OBSERVE RESOLUTION ON DIFFERENT CATEGORIES OF CRIME**

```
df_pd = df_opt1.toPandas()
resolution = list(set(df_pd.Resolution))
resolution
```

```
def resolved(Resolution):
  if Resolution == 'NOT PROSECUTED' or Resolution == 'DISTRICT ATTORNEY REFUSES TO PROSECUTE' or Resolution == 'NONE' or Resolution == 'COMPLAINANT REFUSES TO PROSECUTE':
    return False
  return True

resolved_udf = udf(resolved, BooleanType())
df_q7 = df_opt1.withColumn("Resolved", resolved_udf('Resolution'))
df_q7.createOrReplaceTempView('new_df')
```

```
unResolvedCase = spark.sql("SELECT Category, count(*) as count FROM new_df where Resolved == 'False' and Category <> 'NON-CRIMINAL' GROUP BY Category ORDER BY count DESC")
ResolvedCase = spark.sql("SELECT Category, count(*) as count FROM new_df where Resolved == 'True' and Category <> 'NON-CRIMINAL' GROUP BY Category ORDER BY count DESC")

display(unResolvedCase)
display(ResolvedCase)
```

After categorized types of resolution into resolved and unresolved, looking at the top 5 of each categories, I realized that larceny theft has a large proportion of unsolved cases among total number of cases, yet solved cases has only taken small amount of number among total solved cases. this is similar to veichle theft. 

this find means that stealing is the most serious problem in SF, yet finding ways to resolve is a very important. 

here I suggest that SF downtown area police department and regional government should come up or strict on the laws about stealing related crimes.

```
#which police district has the lowest resolution rate
district_reso = spark.sql("SELECT PdDistrict, Resolution, count(*) as count FROM sf_crime WHERE PdDistrict != 'NA' GROUP BY PdDistrict, Resolution ORDER BY 1, 2")
district_reso.createOrReplaceTempView('district_reso')
district_reso_rate = spark.sql("SELECT PdDistrict, Resolution, sum(count)/sum(sum(count)) over (Partition by PdDistrict) as rate FROM district_reso GROUP BY 1,2 ORDER BY 1, 3 DESC")
display(district_reso_rate.filter(district_reso_rate.Resolution == 'NONE').orderBy(district_reso_rate.rate, ascending = False))
```
